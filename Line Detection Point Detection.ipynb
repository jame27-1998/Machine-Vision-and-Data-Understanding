```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Point Detection
```

```python
src = cv.imread('points.png')
```

```python
plt.imshow(src)
```

```python
src = cv.GaussianBlur(src, (3, 3), 0)
```

```python
src_gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
```

```python
kernel_size = 3
ddepth = cv.CV_16S
dst = cv.Laplacian(src_gray, ddepth, ksize=kernel_size)
plt.imshow(dst)
```

```python
abs_dst = cv.convertScaleAbs(dst)
plt.imshow(abs_dst)
```

```python
sr = cv.imread('roadline.jpg')
```

```python
plt.imshow(sr)
```

```python
#Vertical Line Detection
kernel1 = np.array([[-1, 2, -1],
                    [-1, 2, -1],
                    [-1, 2, -1]])
```

```python
vertical = cv.filter2D(src=sr, ddepth=-1, kernel=kernel1)
```

```python
plt.imshow(vertical)
```

```python
#Horizontal Line Detection
kernel2 = np.array([[-1, -1, -1],
                    [2, 2, 2],
                    [-1, -1, -1]])
```

```python
horizontal = cv.filter2D(src=sr, ddepth=-1, kernel=kernel2)
```

```python
plt.imshow(horizontal)
```

```python
#Edge Detection
```

```python
sr = cv.imread('cat.jpg')
kernel3 = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
```

```python
edges = cv.filter2D(src=sr, ddepth=-1, kernel=kernel3)
```

```python
plt.imshow(edges)
```

```python
kernel4 = np.array([[1, 0, -1],
                    [2, 0, -2],
                    [1, 0, -1]])
```

```python
edges = cv.filter2D(src=sr, ddepth=-1, kernel=kernel4)
```

```python
plt.imshow(edges)
```
