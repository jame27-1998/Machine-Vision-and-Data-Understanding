```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
##reading image
```

```python
img = cv.imread('cat.jpg')
cv.imshow("Cat",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
##updating image
```

```python
imt = cv.imread('cat.jpg',0)
cv.imwrite('greycat.jpg',imt)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## image size,shape,type
```

```python
imt = cv.imread('cat.jpg')
print('Image shape is ',imt.shape)
print('Image size(pixels) is ',imt.size)
print('Image type is ',imt.dtype)
```

```python
## image blending
```

```python
img1 = cv.imread('cat.jpg')
img2 = cv.imread('cat1.jpg')

dst = cv.addWeighted(img1,0.5,img2,0.3,0)
cv.imwrite('blend.jpg',dst)
cv.imshow('dst',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## blur
```

```python
image = cv.imread('cat.jpg')
  
cv.imshow('Original Image', image)
cv.waitKey(0)
  

Gaussian = cv.GaussianBlur(image, (7, 7), 0)
cv.imwrite('blurry.jpg',Gaussian)
cv.imshow('Gaussian Blurring', Gaussian)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# greyscaling image
```

```python
image = cv.imread('cat.jpg')
cv.imshow('Original', image)
cv.waitKey(0)
 

gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
 
cv.imshow('Grayscale', gray_image)
cv.waitKey(0) 
 

cv.destroyAllWindows()
```

```python
# edge detection
```

```python
img = cv.imread('cat.jpg')
edges = cv.Canny(img, 100, 200)
cv.imwrite('result.jpg', edges)

greyscale = cv.imread('result.jpg')
cv.imshow('Grayscale',greyscale)
cv.waitKey(0) 
 

cv.destroyAllWindows()
```

```python
##filter
img = cv.imread('cat.jpg')
# Apply bilateral filter with d = 15,
# sigmaColor = sigmaSpace = 75.
bilateral = cv.bilateralFilter(img, 15, 75, 75)
cv.imshow("filterrd",bilateral)
cv.imwrite('filtered.jpg',bilateral )

cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
##image rotation
rows,cols = img.shape[:2] 
#(col/2,rows/2) is the center of rotation for the image 
# M is the cordinates of the center 
M = cv.getRotationMatrix2D((cols/2,rows/2),90,1) 
dst = cv.warpAffine(img,M,(cols,rows)) 
plt.imshow(dst)
```

```python
#scaling
```

```python
img = cv2.imread('cat.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
```

```python
#translation
```

```python
img = cv2.imread('cat.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst)
```

```python
#rotation
```

```python
rows,cols = img.shape[:2] 
#(col/2,rows/2) is the center of rotation for the image 
# M is the cordinates of the center 
M = cv.getRotationMatrix2D((cols/2,rows/2),90,1) 
dst = cv.warpAffine(img,M,(cols,rows)) 

plt.imshow(dst)
```
